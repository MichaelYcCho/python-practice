```python
# 거듭제곱을 코드로 구현해보자

%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt

# 거듭제곱은 ** 으로 나타낸다.
def my_func(x):
    a = 3
    return x**3

x = np.linspace(0, 2)
y = my_func(x)

plt.plot(x, y)
plt.xlabel("x", size=14)
plt.ylabel("y", size=14)
plt.grid()
plt.show()
```

```python
#제곱근

%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt

# sqrt()는 양의 제곱근을 나타내는 수로, x**(1/2) 와 같은 말이다. 
def my_function(x):
    return np.sqrt(x)

x = np.linspace(0, 9)
y = my_function(x)

plt.plot(x, y)
plt.xlabel("x", size=14, color='white')
plt.ylabel("y", size=14, color='white')
plt.grid()
plt.show()
```

```python
# 다항식

def my_function(x):
    return 3*x**2 + 2*x + 1

x = np.linspace(-2, 2)
y = my_function(x)

plt.plot(x, y)
plt.xlabel("x", size=14, color='white')
plt.ylabel("y", size=14, color='white')
plt.grid()
plt.show()
```
