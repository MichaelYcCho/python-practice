```python
#그래프 그리기 

%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-5, 5)
# x에 2를 곱한값을 y좌표로 한다. 
y = 2 * x

plt.plot(x, y)
plt.show()
```

```python
#그래프 그리기 

%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-5, 5)
# x에 2를 곱한값을 y좌표로 한다. 
y_1 = 2 * x
y_2 = 3 * x

#축의 라벨
plt.xlabel("x value", size=14)
plt.ylabel("y value", size=14)

# 그래프 타이틀
plt.title("First Pyplot Graph")

#그리드 표시
plt.grid()

# 선의 스타일을 지정
plt.plot(x, y_1, label="y1")
plt.plot(x, y_2, label="y2", linestyle="dashed")
#범례(좌측 상단) 표시
plt.legend()

plt.show()
```
