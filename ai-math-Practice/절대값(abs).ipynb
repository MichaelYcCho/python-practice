```python
import numpy as np

x = [-5, 5, -1.28, np.sqrt(5), -np.pi/2]

print(np.abs(x))
```

```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-np.pi, np.pi)
y_sin = np.abs(np.sin(x))
y_cos = np.abs(np.cos(x))

plt.scatter(x, y_sin, label="sin")
plt.scatter(x, y_cos, label="cos")
plt.grid()

plt.show()
```
