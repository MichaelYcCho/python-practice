```python
#난수
# 난수는 0 부터 시작하므로 아래처럼하면 1 ~ 6까지 랜덤하게 출력한다. 

import numpy as np

r_int = np.random.randint(6) + 1
print(r_int)

# rand를 사용하면 0 ~ 1 사이의 소수를 랜덤으로 표시한다. 
r_dec = np.random.rand()
print(r_dec)
```

```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt

n = 1000
x = np.random.rand(n)
y = np.random.rand(n)

plt.scatter(x, y)
plt.grid()
plt.show()
```

```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt

n = 1000
x = np.random.randn(n)
y = np.random.randn(n)

plt.scatter(x, y)
plt.grid()
plt.show()
```
