# 역행렬

어떤 수치와 그의 역수를 곱하면 1이된다. 

\
$$3 * \frac{1}{3} = 1$$

스칼라와 마찬가지로 행렬에도 곱하면 단위행렬이 되는 행렬이 존재한다. 
이를 역행렬이라고 한다. 

$$A^{-1}A = AA^{-1} = E $$

\
\
$$C=\begin{pmatrix}

\end{pmatrix}
$$

\

\end{pmatrix}
$$

\

\end{pmatrix}
$$


#행렬식

행렬에 따라서 역행렬이 존재하지않을수 있는데 이는 행렬식에의해판정된다. 

다음 행렬을 보자
\
$$A=
\begin{pmatrix}
a & b \\
c & d
\end{pmatrix}
$$

행렬식은 |A|또는 det A로 나타낸다

|A| = det A = ad-bc
\

이 행렬식이 0인 경우, 역행렬은 존재하지 않는다. 

\
$$A=
\begin{pmatrix}
1 & 2 \\
3 & 4
\end{pmatrix}
$$
위의 행렬은 (1 * 4 ) - (2 * 3) = -2 이므로 역행렬이 존재한다. 


\
$$A=
\begin{pmatrix}
1 & 2 \\
0 & 0
\end{pmatrix}
$$

해당행렬은 0이므로 조내하지않는다. 

역행렬이 존재하는 경우, 다음 공식에 의해 역행렬을 구할수 있다 .
\

$$A^{-1} = \frac{1}{ad-bc}
\begin{pmatrix}
d & -b \\
-c & a
\end{pmatrix} $$
